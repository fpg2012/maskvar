```python
from datasets.coco_lvis import LvisDataset
from datasets.hqseg44k import HQSeg44KTrainDataset
from utils import resize_longest_side
from models.sam_image_encoder import ImageEncoderViT as SamImageEncoder
from tqdm import tqdm
from maskseg_build_everything import build_cocolvis_dataset
```

```python
train_set, val_set = build_cocolvis_dataset()
```

```python
def count_masks(dataset: LvisDataset | HQSeg44KTrainDataset, splits=4):
    count = [0] * splits
    for i, sample in enumerate(tqdm(dataset)):
        count[i % splits] += len(sample[2])
    return count
```

```python
result = count_masks(train_set, splits=4)
```
